### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

```python
user = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user',delimiter ='|')
```

### Step 3. Assign it to a variable called users.

```python
user.head()
```

### Step 4. Discover what is the mean age per occupation

```python
occupation = user.groupby('occupation')
```

```python
occupation.mean()
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
new = user.groupby(['occupation','gender'])['user_id'].count().reset_index()
new['ratio'] = new.groupby('occupation')['user_id'].apply(lambda x: x/x.sum())
ratio = new[new['gender'] == 'M'].sort_values(by='ratio',ascending=False)[['occupation','ratio']]
ratio.reset_index().drop('index',axis=1)
ratio
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
user.groupby(['occupation','gender'])['age'].aggregate({'min','max'});
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
user.groupby(['occupation','gender'])['age'].mean()
```
